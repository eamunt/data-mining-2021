#2. Làm quen với biểu đồ hộp (Box plot)

```python
#import các thư viện cần thiết
import seaborn as sns

#bước 2: Nạp bộ dữ liệu tips trong thư viện seaborn
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.total_bill.max()
```

```python
tips.total_bill.min()
```

#bước 3a: vẽ biểu dồ hình hộp (box plot) cho thuộc tính total_bill

```python
sns.boxplot(y=tips['total_bill'])
```

#vẽ biểu dồ hình hộp cho thuộc tính total_bill được nhóm theo các ngày trong tuần

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

#Vẽ biểu đồ hình hộp cho thuộc tính total_bill của những khác hút thuốc và không hút thuốc được nhóm theo các ngày trong tuần

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette = 'Set3')
```
