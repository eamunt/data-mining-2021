#5. Vẽ biểu đồ mô tả mối quan hệ giữa size và total_bill trong tập dữ liệu tips của thư viện seaborn

```python
#import thư viện
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

```python
sns.lmplot(x="total_bill", y="size", data = df)
```

#Điều kiện trên smoker

```python
sns.lmplot(x='total_bill', y = 'size', hue='smoker', data=df)
```

#Đổi kiểu của points

```python
sns.lmplot(x = 'total_bill',
           y= 'size',
           hue = 'smoker',
           markers =["o","x"],
           data=df)
```

#Đổi bảng màu khác

```python
sns.lmplot(x="total_bill", 
           y='size', 
           hue='smoker', 
           data = df, 
           palette = 'Set1')
```

#Ánh xạ các Smoker thành màu bằng dict

```python
sns.lmplot(x='total_bill', 
           y='size', 
           hue='smoker', 
           data = df, 
           palette=dict(Yes='g', No="m"))
```

#Chia ra smoker = yes và smoker = no

```python
sns.lmplot(x='total_bill',
           y='size',
           hue ='smoker',
           col='smoker',
           data=df)
```

#Nhiều cột, thay đổi chiều cao và các tỷ lệ co của các khía cạnh

```python
sns.lmplot(x='total_bill',
           y='size',
           hue='day',
           col='day',
           data = df, 
           height=7, 
           aspect=.4,
           x_jitter=.1)
```

#Nhiều hàng

```python
sns.lmplot(x='total_bill',
           y='size',
           col='day',
           hue='day',
           data = df,
           col_wrap=2,
           height=3)
```

#Điều kiện 2 biến sex và time

```python
sns.lmplot(x='total_bill',
           y='size',
           row='sex',
           col='time',
           hue = 'sex', #màu cho sex
           data=df,
           height=3)
```

```python
sns.lmplot(x='total_bill',
           y='size',
           hue="day",
           markers='*',
           data = df)
```

#Sử dụng FacetGrid để tùy chỉnh thêm

```python
g = sns.lmplot(x='total_bill',
                y='size',
                row='sex',
                col='time',
                data = df,
                height=7)

#tùy chỉnh thêm
g.set_axis_labels("total_bill", "size").set(
       xlim=(0, 57),
       ylim=(0, 10), 
       xticks=[10,30,50],
       yticks=[2,6,10]).fig.subplots_adjust(wspace=.02)
```
