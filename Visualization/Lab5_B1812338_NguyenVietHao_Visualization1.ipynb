```python
## buoc #1: import cac thu vien can thiet
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# buoc #2: Nap du lieu: Chieu cao cua 500 khach hang o 1 phong tap gym
df = pd.read_csv("https://raw.githubusercontent.com/ltdaovn/dataset/master/gym_height.csv", index_col = 0)
df.head()
```

#Biểu đồ thứ I height được biểu diễn bằng 10 cột, biểu đồ thứ II height được biểu diễn bằng 20 cột
#Bins = 20: chia thành 20 cột, không khai báo thì mặc định là 10 cột

```python
## buoc #3: ve histogram cho thuoc tinh chieu cao
df.hist(column = 'height', grid = False)
df.hist(column = 'height', grid = False, bins= 20)
```

### a. Vẽ histogram cho dữ liệu chiều cao của từng nhóm khách hàng (nam/nữ).

```python
df.hist(column = 'height', by="sex", grid = False, figsize = (14,7), bins=20)
```

### b. Vẽ histogram chiều cao của nam và nữ trên cùng 1 biểu đồ.

```python
f= df.height[df['sex'] == "f"]
m= df.height[df['sex'] == "m"]
plt.hist([f,m], bins = 20, label=['f','m'])
plt.legend()
plt.show()
```
