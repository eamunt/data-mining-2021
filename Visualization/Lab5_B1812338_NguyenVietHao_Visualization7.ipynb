#7. Hãy phân tích mối tương quan giữa các thuộc tính trong tập dữ liệu USArrests

```python
#import các thư viện cần thiết
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#nạp dữ liệu
df = pd.read_csv('https://raw.githubusercontent.com/ltdaovn/dataset/master/USArrests.csv')
# Chuẩn hóa tên cột
df.rename(index=str ,columns={"Unnamed: 0" :"Bang"}, inplace=True)
df
```

```python
df.count()
```

#Bắt giữ murder(giết người) và Urban Population(dân số thành thị)

```python
sns.lmplot(x='Murder', 
           y='UrbanPop',
           data =df)
#không có thay đổi gì nhiều giữa số lượng Murder bị bắt so với UrbanPop
```

#Bắt giữ Tấn công(Assault) và Urban Population(dân số thành thị)

```python
sns.lmplot(y='UrbanPop',
           x='Assault',
           data = df)
```

# => Việc bắt giữ giết người(Murder) và tấn công(Assault) dường như không có mối tương quan nào đối với người dân thành thị(UrbanPop)

#==================

#Bắt giữ Rape với UrbanPop

```python
sns.lmplot(x='Rape', 
           y='UrbanPop', 
           data = df)
#biểu đồ cho thấy số lượng Rape bị bắt giữ cao ở các nơi có UrbanPop lớn
#Rape và UrbanPop có mối tương quan cao với nhau.
```
