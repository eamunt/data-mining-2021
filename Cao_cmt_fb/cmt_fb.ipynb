```python
import random
#khai báo biến browser
browser = webdriver.Chrome('chromedriver',chrome_options=chrome_options)

#mở trang post cần crawling cmt
browser.get("https://www.facebook.com/daa.ctu.edu.vn/posts/3223134181068646")

time.sleep(random.randint(5,10))


#tài khoản mật khẩu này em dùng làm acc chính nên em đã đổi tài khoản mật khẩu ạ.
#em chỉ dùng để chạy đc rồi, e đã đổi vì là acc chính của em ạ. em bị ban mấy nick rồi
username = browser.find_element_by_id('email')
username.send_keys("nhaptaikhoan")

passw = browser.find_element_by_id('pass')
passw.send_keys("nhapmatkhau")

#đăng nhập
log = browser.find_element_by_id('loginbutton')
log.click()



#click bình luận

data=[]

#click xem các bình luận trước
sleep(random.randint(5,10))

while True:
    try:
      WebDriverWait(browser, 20).until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div[4]/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div[4]/div/div/div[2]/div[2]/div[1]/div[2]/span/span"))).click()
      sleep(random.randint(5,10))
    except:
      break
    sleep(random.randint(5,10))


sleep(random.randint(5,10))
#lấy các thẻ div có thuộc tính có 
cmt_list = browser.find_elements_by_xpath("//div[@class='tw6a2znq sj5x9vvc d1544ag0 cxgpxx05']")

#lặp trong tất cả các cmt để lấy tên và nội dung
for cm in cmt_list:
    usr = cm.find_element_by_class_name("nc684nl6")
    try:
        content = cm.find_element_by_class_name("ecm0bbzt.e5nlhep0.a8c37x1j")
        print("--", usr.text,":",content.text)
        data.append(usr.text)
        data.append(content.text)
    except:
        print("--", usr.text)
        data.append(usr.text)

#chờ 5 giây
sleep(random.randint(5,10))


#đóng browser
df = pd.DataFrame(data)
df.to_csv("cmt_fb1.csv", index=False)
#đóng browser
browser.close()
```
